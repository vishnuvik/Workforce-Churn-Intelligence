# Q-Q plot for data distribution

```python
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt


# Generate the Q-Q plot for the 'number_project' feature
feature = df['average_montly_hours']

# Create Q-Q plot
plt.figure(figsize=(8, 6))
stats.probplot(feature, dist="norm", plot=plt)

# Customize the plot
plt.title('Q-Q Plot of number_project')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')

# Show the plot
plt.show()
```

```python
feature = df['average_montly_hours']

# Apply reciprocal transformation
reciprocal_feature = 1 / feature

# Create Q-Q plot
plt.figure(figsize=(8, 6))
stats.probplot(reciprocal_feature, dist="norm", plot=plt)

# Customize the plot
plt.title('Q-Q Plot of Reciprocal Transformed Average Monthly Hours')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')

# Show the plot
plt.show()
```

```python
feature = df['average_montly_hours']

# Apply square root transformation
sqrt_feature = np.sqrt(feature)

# Create Q-Q plot
plt.figure(figsize=(8, 6))
stats.probplot(sqrt_feature, dist="norm", plot=plt)

# Customize the plot
plt.title('Q-Q Plot of Square Root Transformed Average Monthly Hours')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')

# Show the plot
plt.show()
```

```python
feature = df['average_montly_hours']

# Apply log transformation
# Adding 1 to avoid log(0) which is undefined
log_feature = np.log(feature + 1)

# Create Q-Q plot
plt.figure(figsize=(8, 6))
stats.probplot(log_feature, dist="norm", plot=plt)

# Customize the plot
plt.title('Q-Q Plot of Log-transformed Average Monthly Hours')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['average_montly_hours'], bins=30, edgecolor='k', alpha=0.7)

# Customize the plot
plt.title('Histogram of Average Monthly Hours')
plt.xlabel('Average Monthly Hours')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
feature = df['average_montly_hours']

skewness = stats.skew(feature)
kurtosis = stats.kurtosis(feature)

print(f'Skewness of average_monthly_hours: {skewness}')
print(f'Kurtosis of average_monthly_hours: {kurtosis}')
```

```python
feature = df['satisfaction_level']

skewness = stats.skew(feature)
kurtosis = stats.kurtosis(feature)

print(f'Skewness of average_monthly_hours: {skewness}')
print(f'Kurtosis of average_monthly_hours: {kurtosis}')
```
